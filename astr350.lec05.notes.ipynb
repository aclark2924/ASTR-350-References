## Changes in coordinate angles vs. angular separation angles

```python
# one point on the celestial sphere
ra1 = 185 * u.deg
dec1 = 60 * u.deg

# another point with a different R.A.
ra2 = ra1 + 15 * u.deg
dec2 = dec1
```

```python
# what is the angular separation between point 1 and point 2?
import numpy as np
(ra2 - ra1) * np.cos(dec1)
```

### What if both R.A. and Dec. change?

```python
ra1 = 185 * u.deg
dec1 = 60 * u.deg
ra2 = ra1 - 30 * u.deg
dec2 = dec1 - 40 * u.deg
```

```python
# approximate formula for small angular separations
dra = (ra2 - ra1) * np.cos(dec1)
ddec = dec2 - dec1
ang_sep = np.sqrt( dra**2 + ddec**2 ) #.to(u.arcmin)
print(f"angular separation is {ang_sep:0.3f}")
```

```python
# exact angular separation 
from astropy.coordinates import SkyCoord

pos1 = SkyCoord(ra1, dec1)
pos2 = SkyCoord(ra2, dec2)
ang_sep = pos1.separation(pos2).to(u.deg)
#ang_sep = pos1.separation(pos2).to(u.arcmin)
print(f"angular separation is {ang_sep:0.3f}")
```
