## Frequentist vs. Bayesian Statistics

Frequentist Approach:
$$ P \left( Data \,\middle|\, Model \right) $$

Bayesian Approach:
$$ P \left( Model \,\middle|\, Data \right) $$

## Probability (Kolmogorov Axioms)

![](media/prob_sets.png)

Let $\Omega$ be a collection of possible events including $A$ and $B$ (e.g., $A, B \in \Omega$).

**Axiom 1:** $P(E) \geq 0$ for all $E$

**Axiom 2:** $P(\Omega)=1$

**Axiom 3:** $P(\cup_i^\infty E_i) = \sum_i^\infty P(E_i)$ if $E_1, E_2, \ldots$, are mutually exclusive.



## Conditional Probability

"What are the chances of A given B"

![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/86/Venn_Pie_Chart_describing_Bayes%27_law.png/586px-Venn_Pie_Chart_describing_Bayes%27_law.png)

$$ P \left( A \,\middle|\, B \right)  = {P(A \cap B) \over P(B)}$$

where $A \cap B$ is the intersection of A and B.

re-aranging:
$$P \left( A \,\middle|\, B \right) P(B) = P(A \cap B)  $$

Since

$$P(A \cap B) = P(B\cap A)$$

then 

$$P \left( A \,\middle|\, B \right) P(B) =  P \left( B \,\middle|\, A \right) P(A) $$

and

$$ P \left( A \,\middle|\, B \right)  = {P \left( B \,\middle|\, A \right) P(A) \over P(B)}$$

## Bayes' Theorem

$$ P \left( A \,\middle|\, B \right)  = {P \left( B \,\middle|\, A \right) P(A) \over P(B)}$$

Changing the notation, we can write:

$$ P \left( Model \,\middle|\, Data \right)  = {P \left( Data \,\middle|\, Model \right) P(Model) \over P(Data)}$$

where:
- $P \left( Data \,\middle|\, Model \right)$ is the likelihood, $\mathcal{L}$
- $P(Model)$ is the prior on the model
- $P(Data)$ esentially serves to normalize the numerator
- $P \left(Model \,\middle|\, Data \right)$ is the posterior probability distribution
  
Note that the symbol, $\theta$ is often used to represent the model (parameters) and $I$ is often used to represent the $Data$ (and/or other information). So the equation above is often written: 

$$ P \left( \theta \,\middle|\, I \right)  = {P \left( I \,\middle|\, \theta \right) P(\theta) \over P(I)}$$

### Key difference between Frequentist and Bayesian approachies:

* Bayesian approach allows for **prior knowledge** to be included
* Bayesian approach produces a **posterior** probability that requires data to calculate
