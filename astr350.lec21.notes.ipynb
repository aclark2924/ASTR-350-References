## Fourier Decomposition

Any periodic function, $f(x)$, can be expressed as a sum of sine and cosine waves of various amplitudes and frequencies as follows:

$$f(x) = {1 \over 2}a_0 + \sum_{n=1}^{\infty}a_n \cos(nx) + \sum_{n=1}^{\infty}b_n \sin(nx)  $$

(see http://mathworld.wolfram.com/FourierSeries.html)

## Aliasing

If you sample at a given rate there will be multiple signal frequencies that can match the sampled amplitudes.

## Nyquist Sampling theorm

A signal sampled on **regular intervals** $1 / f_s$ apart contains all information necessary to reconstruct frequencies larger than $f > 2 f_s$.
