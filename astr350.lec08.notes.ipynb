## Moments of a Sample

* given a sample of $N$ data points, $\{x_0, x_1, ... , x_{N-1}\}$:

$1^{\rm st}$ moment, the **Mean**
$$ \bar{x} = {1 \over N}\sum_{i=0}^{N-1} x_i$$

$2^{\rm nd}$ moment, the **Variance**
$$ \sigma^2 = {1 \over N-1}\sum_{i=0}^{N-1} (x_i - \bar{x})^2$$

(note the "Standard Deviation", $\sigma$, is the square root of the variance)

$3^{\rm rd}$ moment, the **Skewness**
$$ \gamma = {1 \over N}\sum_{i=0}^{N-1} \left({x_i - \bar{x} \over \sigma}\right)^3$$

$4^{\rm th}$ moment, the **Kurtosis**
$$ {\rm Kurt} = {1 \over N}\sum_{i=0}^{N-1} \left({x_i - \bar{x} \over \sigma}\right)^4 - 3$$

## Gaussian Distribution

$$ G_{\mu \sigma}(x) = { 1 \over \sigma \sqrt{2\pi}} e^{-{(x - \mu)^2 \over 2\sigma^2}}$$

where:
  * $x$ is a real number (positive or negative)
  * $\mu$ is the expectation value (center of the distribution)
  * $\sigma$ is the standard deviation of the distribution

## Gaussian Function in Python

```python
def gaussian(x, mu=0, sigma=1):
    norm = sigma * np.sqrt(2 * np.pi)
    exp = np.exp( -(x - mu)**2 / (2 * sigma**2) )
    return exp / norm
```

```python
def plot_gauss(mu=0, sigma=1):
    # plot a Gaussian
    x = np.linspace(mu - 5 * sigma, mu + 5 * sigma, 100)
    y = gaussian(x, mu, sigma)
    plt.plot(x, y, lw=8)
    plt.grid(ls='--')
    
    # show the 1-sigma width
    x0 = mu
    x1 = mu + sigma
    y1 = gaussian(x1, mu, sigma)
    plt.plot([x0, x1], [y1, y1], lw=3, label='$\sigma$')

    # show the full width at half max (FWHM)
    fwhm = 2.355 * sigma
    xa = mu - fwhm / 2
    xb = mu + fwhm / 2
    yb = gaussian(xa, mu, sigma)
    plt.plot([xa, xb], [yb, yb], lw=3, label='FWHM')

    plt.legend()
```

```python
plot_gauss()
```

## Confidence Intervals

```python
def plot_ci(nsigma=1, mu=0, sigma=1):
    # plot the Gaussian
    x = np.linspace(mu - 5 * sigma, mu + 5 * sigma, 100)
    y = gaussian(x, mu, sigma)
    plt.plot(x, y, lw=8)
    plt.grid(ls='--')

    # get integrated area under the curve within nsigma from the mean
    x0 = mu - nsigma * sigma
    x1 = mu + nsigma * sigma
    xs = np.linspace(x0, x1, 1000)
    ys = gaussian(xs, mu, sigma)
    frac = np.trapz(ys, xs)

    # annotate the plot
    plt.fill_between(xs, ys, color='0.8')
    plt.annotate("$\pm{}\sigma$".format(nsigma), xy=(mu, y.max()/2), ha='center', fontsize=18)
    plt.annotate("{:.2f}%".format(100 * frac), xy=(mu, y.max()/2.1), ha='center', va='top', fontsize=18)
    plt.axvline(mu - nsigma * sigma, c='r')
    plt.axvline(mu + nsigma * sigma, c='r')
    plt.xlabel('x')
    plt.ylabel('Relative Probability')
```

```python
plot_ci()
```
