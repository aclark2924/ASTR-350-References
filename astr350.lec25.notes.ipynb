## Interstellar absorption

With no absorption, recall that observed magnitudes, $m$, and absolute magnitudes, $M$, are related by: 
$$ m = M + 5 \log \left({D \over 10 {\rm pc}} \right)$$

Lets define a [distance modulus](https://en.wikipedia.org/wiki/Distance_modulus), $\mu$, to be the change in magnitudes of a source due to distance:
$$\mu = 5 \log{\left({D \over 10 {\rm pc}} \right)}$$

The observed magnitude, $m_X$, in band $X$ can be computed from the absolute magnitude, $M_X$, the distance modulus, $\mu$, and the interstellar absorption, $A_X$ as follows:
$$m_X = M_X + \mu + A_X$$

For example, the observed $V$-band magnitude is given by:
$$m_V = M_V + \mu + A_V$$

(note we often write $V$ for the observed $V$-band magnitude instead of $m_V$)

## Interstellar reddening

The color excess, $E(B-V)$, is the difference between the observed $B-V$ color of an object and its intrinsic, $(B-V)_0$ color:

$$E(B-V) = (B - V) - (B - V)_0$$

Another way to think of this is:
$$E(B-V) = A_B - A_V $$

## Relation between interstellar reddening and interstellar absorption

- Interstellar dust will make distant objects appear both fainter and redder, but absorption and redding are correlated
- Thus if we know the intrinsic color of a source and measure its color excess, we can estimate the absorption

$A_V$ and $E(B-V)$ are related by:
$$R_V =  {A_V \over E(B-V)}$$
where $R_V$, the ratio between the total extinction and the selective extinction, is typically about 3.1.

## Extinction Law

![Cardelli law](media/ccm89.png)

Dust models, such as the [Cardelli law](http://adsabs.harvard.edu/abs/1989ApJ...345..245C), can be used to predict the absorption, $A_{\lambda}$, at other wavelengths relative to the absorption, $A_V$, in the V-band. 

For example, here are the coefficients in some other bands for a CCM89 law with $R_V=3.1$:

$$
\begin{align}
A_U  & =  1.664 \, A_V\\
A_B  & =  1.321 \, A_V\\
A_R  & =  0.819 \, A_V\\
A_I  & =  0.594 \, A_V\\
\end{align}
$$

(see Table 6 in [Schlegel, Finkbeiner & Davis, 1998](https://ui.adsabs.harvard.edu/abs/1998ApJ...500..525S/abstract))
