## Review: Least Squares Fitting

```python
data = np.genfromtxt('media/xy.dat', names='x, y')
plt.plot(data['x'], data['y'], 'ro');
plt.grid();
```

```python
# fit a line in the least-squares sense
X = np.matrix( [[xi, 1] for xi in data['x']] )
Y = np.matrix( [[yi] for yi in data['y']] )
p = (X.T * X).I * X.T * Y
m, b = p.A1

# get the uncertainty
modelY = X * p
dY = Y - modelY
sample_var = dY.var(ddof=p.shape[0])
pvar = sample_var * np.diagonal( (X.T * X).I )
msig, bsig = np.sqrt(pvar)

print(f"slope is {m:.1f} +/- {msig:.1f}")
```

## Jackknife

```python
def jackknife_dist(sample, func):
    dist = np.zeros(sample.size)
    wuse = np.ones(sample.size, dtype=bool)
    for i in range(sample.size):
        wuse[:] = True
        wuse[i] = False
        dist[i] = func(sample[wuse])
    return dist
```
