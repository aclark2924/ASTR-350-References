## Interstellar reddening

Recall from [Lecture 25](astr350.lec25.notes)

The color excess, $E(B-V)$, is the difference between the observed $B-V$ color of an object and its intrinsic, $(B-V)_0$ color:

$$E(B-V) = (B - V) - (B - V)_0$$

## Relation between interstellar reddening and interstellar absorption

- Interstellar dust will make distant objects appear both fainter and redder, but absorption and redding are correlated
- Thus if we know the intrinsic color of a source and measure its color excess, we can estimate the absorption

The observed magnitude, $m_X$, in band $X$ can be computed from the absolute magnitude, $M_X$, the distance modulus, $\mu$, and the interstellar absorption, $A_X$ as follows:
$$m_X = M_X + \mu + A_X$$

$A_V$ and $E(B-V)$ are related by:
$$R_V =  {A_V \over E(B-V)}$$
where $R_V$, the ratio between the total extinction and the selective extinction, is typically about 3.1.

[Dust models](http://adsabs.harvard.edu/abs/1989ApJ...345..245C) can be used to predict the absorption, $A_{\lambda}$, at other wavelengths relative to the absorption, $A_V$, in the V-band. For example, here are the coefficients in the Gaia bands:

$$
\begin{align}
A_G  & = & 0.86073 \, A_V\\
A_{G\_BP}  & = & 1.06244 \, A_V\\
A_{G\_RP}  & = & 0.63984 \, A_V\\
\end{align}
$$
