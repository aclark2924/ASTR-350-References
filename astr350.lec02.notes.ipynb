## Example: get a local copy of the course repositories (if you have not already)

Set up `git` (if necessary): <br>
- `git config --global user.name "Robert Quimby"`
- `git config --global user.email "rquimby@sdsu.edu"`

GitHub now requires "tokens" for command line access
- https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/creating-a-personal-access-token

First set up a token on GitHub if you have not already:
- https://github.com/settings/tokens
- or, in any GitHub page, click on your Profile icon (upper right) → Settings → Developer settings (bottom of left panel) → Personal access tokens

At the (unix) command line `cd` to a directory where you want to store the lecture notes (e.g., your home directory), then type:<br>
```git clone https://github.com/SDSU-astr350-2026-spring/lectures.git```

### Store your GitHub token for later use
- `git config --global credential.helper store`

Execute similar commands for syllabus, homework, and tutorials repositories

- `git clone https://github.com/SDSU-astr350-2026-spring/syllabus.git`
- `git clone https://github.com/SDSU-astr350-2026-spring/homework.git`
- `git clone https://github.com/SDSU-astr350-2026-spring/tutorials.git`
