## Galaxy Spectra

```python
# load SDSS spectrum
spec_url = 'http://das.sdss.org/spectro/1d_23/0302/1d/spSpec-51688-0302-325.fit'
flux = fits.getdata(spec_url)[0]

# use header WCS to calculate wavelengths for each spectral bin
head = fits.getheader(spec_url)
wav = 10**(head['crval1'] + head['cd1_1'] * np.arange(flux.size))
```

```python
# plot
plt.plot(wav, flux, lw=2)
plt.grid()
```

## Redshift

- [lengthy Wikipedia article on redshift](https://en.wikipedia.org/wiki/Redshift)

Lets define:
- $\lambda_{\rm rest}$ to be the *rest* wavelength
- $\lambda_{\rm obs}$ to be the *observed* wavelength
- $z$ to be the "redshift" (in dimensionless units)

The relationship between these three quantities is:
$$\lambda_{\rm obs} = \lambda_{\rm rest} (1 + z)$$

### What is the relationship between redshift, $z$, and velocity, $v$?

$$(1 + z) = \sqrt{ 1 + v/c \over 1 - v/c }$$

for small $v$, this is approximately $z \approx v/c$
