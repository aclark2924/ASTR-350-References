## Types of brackets

- **parenthesis** are used for: 
  - precedence: `2 * (1 + 5)`
  - tuples: `(1, 4, 'apple')`
  - function calls: `print('hello')`
- **square brackets** are used for:
  - lists: `[1, 4, 'apple']`
  - indexing: `a[5]`
  - slicing: `a[1:7]`
- **curly braces** are used for:
  - dictionaries: `{'a': 'apple', 'b': 'ball', 'c': 3e10}`
  - sets: `{1, 1, 1, 3, 4, 1, 3, 3, 1} == {1, 3, 4}` 
  - fstrings: `f"x is {x}"`
