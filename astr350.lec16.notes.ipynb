![](http://hyperphysics.phy-astr.gsu.edu/hbase/solar/imgsol/eclip.png)

Ecliptic latitude, $\beta$, and Ecliptic longitude, $\lambda$, have the same origin as Equitorial coordinates (namely the direction to the Vernal equinox), but Ecliptic coordinates are rotated so that the latitudinal coordinate ($\beta$) lines up with the Earth's orbital plane.
