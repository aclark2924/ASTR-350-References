## Aperture Photometry

- sum up all the counts received within some region of the image
- remove from this sum the contribution from background light
- use statistics and uncertainty propagation to determine the uncertainty in the derived source counts

## Circular Aperture Photometry

- choose a radius, r
- choose an (x, y) position on the image
- sum the counts within r of (x, y)
- remove the estimated background contribution from this sum
  - typically we get the background estimate from an annulus around the source

## FWHM for different stars

```python
# bright star position
x1, y1 = 569.42, 424.33

# faint star position
x2, y2 = 663.45, 583.95
```

```python
plt.figure(figsize=(7,7))
plt.imshow(image, origin='lower', cmap='gray', vmin=1400, vmax=1600);
plt.plot(x1, y1, marker='o', mfc='none', c='r', ms=15)
plt.plot(x2, y2, marker='o', mfc='none', c='y', ms=15);
```

```python
def get_profile(image, x, y, margin=10):
    i, j = int(x), int(y)
    stamp = image[j - margin : j + margin + 1, i - margin : i + margin + 1]
    
    profile = stamp.sum(axis=1)
    profile -= profile.min()
    profile /= profile.max()
    
    axis = np.arange(profile.size) - y + j - margin
    
    return axis, profile
```

```python
# get the profiles for each star
axis1, profile1 = get_profile(image, x1, y1)
axis2, profile2 = get_profile(image, x2, y2)
```

```python
plt.plot(axis1, profile1)
plt.grid();
```

## Recall: converting counts to magnitudes

$$m = m_{zp} - 2.5\log_{10}(C)$$
where:
- $m$ is the magnitude of the source
- $m_{zp}$ is the image zeropoint
- $C$ is the counts (or count rate) from the source

Relation between uncertainty in counts and uncertainty in magnitudes:
$$ \sigma_m = {2.5 \over \ln(10)}{\sigma_C \over C} $$
where:
- $\sigma_m$ is the magnitude uncertainty
- $\sigma_C$ is the counts (or count rate) uncertainty

The magnitude difference between two stars is an expression of the ratio of their fluxes on a log scale:

$$ m_1 - m_2 = -2.5 \log_{10} \left( {f_1 \over f_2} \right)$$

Thus you need not measure the total flux from either object to determine their magnitude difference. Rather, you simply need to measure the same fraction of light from each object (thus preserving the flux ratio).
