## Contributions to raw CCD pixel values

```python
npixels = 25
gain = 2.1 # e- / ADU
read_noise = 4.6 # e-

# fixed components (e-)
pix = np.arange(npixels)
photons = ???? * np.ones(npixels)
np.random.seed(42)
static_bias = np.random.uniform(????, size=npixels)
flat_field = np.random.normal(????, 0.05, size=npixels)
```

```python
# random components (e-)
np.random.seed()
floating_bias = ???? + np.random.normal(????)
poisson_noise = np.random.poisson(photons * flat_field) - photons * flat_field
noise = poisson_noise + np.random.normal(0, read_noise, size=npixels)

# total (ADU)
counts = (floating_bias + static_bias + photons * flat_field + noise) / gain
counts
```

```python
# show the contributions to each pixel (in ADU)
plt.bar(pix, floating_bias / gain)
plt.bar(pix, static_bias / gain, bottom=floating_bias / gain)
plt.bar(pix, flat_field * photons / gain, bottom=(floating_bias + static_bias) / gain)
plt.bar(pix, counts, color='k', fill=False)
plt.grid();
```

```python
# recover the number of incident photons
overscan = floating_bias / gain
bias = static_bias / gain
measured_photons = (counts - overscan - bias) * gain / flat_field
plt.bar(pix, measured_photons, color='C2')
plt.grid();
```
